# Update Supply Chain Data

**Use when:** Data files updated in `/Files/data/supplychain/` but schema unchanged

## Process Overview
- **Truncate**: Clear existing data, keep table structure
- **Load Data**: Reload fresh data from CSV files into lakehouse tables

## Prerequisites
- CSV files must be present in `/Files/data/supplychain/` folder
- Tables must already exist in the lakehouse
- ISO date format required (YYYY-MM-DD)

```python
# Configuration - Target schema and tables to refresh
SCHEMA_NAME = "supplychain"  # Target schema in lakehouse
TABLES = ["Suppliers", "ProductSuppliers", "SupplyChainEvents", "SupplyChainEventImpacts"]  # Tables to refresh
```

```python
for table in TABLES:
    try:
        spark.sql(f"TRUNCATE TABLE {SCHEMA_NAME}.{table}")
        print(f"✅ Truncated {table}")
    except Exception as e:
        print(f"⚠️ Skipped {table} (truncate not supported)")
```
