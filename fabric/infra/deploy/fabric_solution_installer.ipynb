## 🚀 Deployment

```python
%pip install fabric-launcher --quiet
import notebookutils
notebookutils.session.restartPython()
```

```python
from fabric_launcher import FabricLauncher

launcher = FabricLauncher(notebookutils,
    api_root_url = "https://api.fabric.microsoft.com" #Default is https://api.fabric.microsoft.com, but may vary depending on your environment
    )


# ── GitHub source settings ────────────────────────────────────────────────────
GITHUB_OWNER     = "DocGailZhou"
GITHUB_REPO      = "solutionroot"
GITHUB_BRANCH    = "main"
GITHUB_FABRIC_WORKSPACE_PATH = "fabric/fabric_workspace"  # Folder inside the repo where the fabric solution is synched

# ── Lakehouse data settings ─────────────────────────────────────────
LAKEHOUSE_NAME = "fabriciq_team_lake"         # Lakehouse in the workspace where the ingested CSV files will be landed
GITHUB_INFRA_DATA_PATH = "fabric/infra"  # Folder inside the repo containing the data to be ingested in the Lakehouse
DATA_FOLDERS={
    f"{GITHUB_INFRA_DATA_PATH}/data": "/data",                                        # Key: folder inside the GitHub repo where the source CSV files are located; Value: target folder in the Lakehouse where the CSV files will be landed
    f"{GITHUB_INFRA_DATA_PATH}/data_dataload_20260316": "/data_dataload_20260316"     # Key: folder inside the GitHub repo where the source CSV files are located; Value: target folder in the Lakehouse where the CSV files will be landed
    }
```

```python
# Deploy solution with data folders
launcher.download_and_deploy(
    repo_owner=GITHUB_OWNER,
    repo_name=GITHUB_REPO,
    branch = GITHUB_BRANCH,
    workspace_folder=GITHUB_FABRIC_WORKSPACE_PATH,
    allow_non_empty_workspace=True,
    item_type_stages = [["Lakehouse", "Ontology"],["Notebook"]],
    validate_after_deployment=True,
    generate_report=False,
    data_folders=DATA_FOLDERS,
    lakehouse_name=LAKEHOUSE_NAME
)
```

## ✅ Automated Post-Deployment Tasks

```python
#Import required libraries
import sempy.fabric as fabric
from fabric_launcher import move_item_to_folder

# Initialize Fabric client and workspace
client = fabric.FabricRestClient()
workspace_id = fabric.get_workspace_id()
```

```python
# Run a notebook to: Create tables
result = launcher.run_notebook_synchronous(
    notebook_name="main_pipeline",
    parameters={},
    timeout_seconds=3600
)
```

```python
# Move current notebook to the notebooks folder
move_item_to_folder(
        item_name=notebookutils.runtime.context["currentNotebookName"],
        item_type="Notebook",
        folder_name="notebooks",
        workspace_id=workspace_id,
        client=client
    )
print("✅ Current notebook moved successfully!")
```
